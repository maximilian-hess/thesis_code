```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def grover_prob(p_0, k):
    return np.sin((2*k+1)*np.arcsin(np.sqrt(p_0)))**2

def randmomized_prob(p_0, k):
    return (1/k)*sum([grover_prob(p_0, l) for l in range(k)])
```

```python
# basic probabilities

p_0 = 2**(-10)
theta = np.arcsin(np.sqrt(p_0))
k_opt = int(np.floor(np.pi/(4*theta)))

grover_iterations = [i for i in range(100)]
grover_probs = [grover_prob(p_0,k) for k in grover_iterations]

fig, ax = plt.subplots()
plt.plot(grover_iterations, grover_probs)

plt.vlines(x=k_opt, ymin=0, ymax=grover_probs[k_opt], color="red")
ax.text(27, 0.1, f"$k^*={k_opt}$")
plt.title(r"Probability to measure a good state")
plt.xlabel("Number of Grover iterations")
plt.ylabel("Probability to measure a good state")
plt.savefig("grover_probs.pdf")
plt.show()
```

```python
# randomized protocol

p_0 = 2**(-10)
cutting_point = 10

grover_iterations = [i for i in range(100)]
grover_probs = [grover_prob(p_0,k) for k in grover_iterations]

fig, ax = plt.subplots()
plt.plot(grover_iterations, grover_probs)
plt.plot(grover_iterations[:cutting_point], grover_probs[:cutting_point], color="red")
plt.vlines(x=cutting_point-1, ymin=0, ymax=grover_probs[cutting_point-1], color="red")
plt.title(f"Average probability to measure a good state, $K={cutting_point}$")
plt.xlabel("Number of Grover iterations")
plt.ylabel("Probability to measure a good state")
ax.fill_between(x=grover_iterations[:cutting_point], y1=[0]*cutting_point, y2=grover_probs[:cutting_point], color="red",alpha=0.2)
ax.text(43, 0.4, f"$p_K=${np.round(randmomized_prob(p_0,cutting_point), decimals=2)}")
plt.savefig(f"avg_prob_K={cutting_point}.pdf")
plt.show()
```
