```python
import numpy as np
from sources.TSP import TSP
from sources.TSPsampler import TSP_sampler
import json
```

```python
# READ INSTANCES

with open("data/tsp_instances.json", "r", encoding="utf-8") as f:
	instances = json.load(f)
```

```python
# TESTS FOR K-OPT WITH EDGE REPLACEMENT

k = 3
sizes = [n for n in range(16,31)]

solutions = {}
for n in sizes:
    solutions[n] = {}
    for i in range(5):
        print(f"Size {n}, instance {i}")
        solutions[n][i] = {}
        matrix = instances[str(n)][str(i)]
        tsp = TSP(matrix)
        sampler = TSP_sampler(tsp)
        greedy_sol, greedy_val = tsp.solve_greedy()
        T_max = n**4
        sol, val, iterations = sampler.sample_k_opt(initial_reference=greedy_sol, k=k, T_max=T_max)
        
        solutions[n][i]["sol"] = sol
        solutions[n][i]["val"] = val
        solutions[n][i]["iterations"] = iterations
        solutions[n][i]["greedy_val"] = greedy_val


with open(f"results/tsp_results_{k}-opt.json", "w") as f:
    json.dump(solutions, f)
```

```python
# TESTS FOR K-OPT WITH VERTEX REPLACEMENT


sizes = [n for n in range(16,31)]

solutions = {}
for n in sizes:
    solutions[n] = {}
    for k in range(2,int(n/2)):
        solutions[n][k] = {}
        for i in range(5):
            print(f"Size {n}, instance {i}, k={k}")
            solutions[n][k][i] = {}
            matrix = instances[str(n)][str(i)]
            tsp = TSP(matrix)
            sampler = TSP_sampler(tsp)
            greedy_sol, greedy_val = tsp.solve_greedy()
            T_max = n**4
            sol, val, iterations = sampler.sample_node_replacement(initial_reference=greedy_sol, k=k, T_max=T_max)
            
            #solutions[n][i]["opt_val"] = opt_val
            solutions[n][k][i]["sol"] = sol
            solutions[n][k][i]["val"] = val
            solutions[n][k][i]["iterations"] = iterations
            solutions[n][k][i]["greedy_val"] = greedy_val


with open(f"results/tsp_results_node_replacement.json", "w") as f:
    json.dump(solutions, f)
```
